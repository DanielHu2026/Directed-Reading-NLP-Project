# 0 — Setup & Config Validation

Loads `configs/config.yaml`, resolves key paths, checks expected inputs, and ensures output folders exist.

```python
from pathlib import Path
import sys

REPO_ROOT = Path.cwd()

sys.path.insert(0, str(REPO_ROOT / "src"))
from l2affect.utils.config import load_config, resolve, repo_root
```

```python
cfg = load_config(REPO_ROOT / ".." / "configs" / "config.yaml")
cfg
```

```python
from pathlib import Path

def ensure_dir(p: Path):
    p.mkdir(parents=True, exist_ok=True)
    return p

paths = cfg["paths"]
inputs = cfg["inputs"]

raw_dir = ensure_dir(resolve(paths["raw_dir"]))
processed_dir = ensure_dir(resolve(paths["processed_dir"]))
fig_dir = ensure_dir(resolve(paths["reports_figures_dir"]))
tbl_dir = ensure_dir(resolve(paths["reports_tables_dir"]))

print("Repo root:", repo_root())
print("Raw dir:", raw_dir)
print("Processed dir:", processed_dir)
print("Reports/figures:", fig_dir)
print("Reports/tables:", tbl_dir)

print("\nExpected inputs:")
for k,v in inputs.items():
    print(f"  {k}: {resolve(v)}")
```
